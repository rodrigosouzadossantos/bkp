# KR1 – Registrar todas novas demandas no GitHub Project

## Descrição

Este KR visa garantir que todas as novas demandas sejam registradas via Issue e vinculadas ao GitHub Project para rastreabilidade e mensuração.

### Entregável

- Lista de novas Issues padronizadas
- Link das Issues no GitHub Project

### Critérios de aceite

- Todas as novas demandas criadas como Issue
- Issue vinculada a Project correto
- Responsável e status preenchidos
- Labels OKR aplicados
