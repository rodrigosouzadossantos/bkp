## Objetivo:
Notebook para fazer a inferência do modelo de qualidade de imagem, em um conjunto de imagens de coral-sol

## Importações Necessárias

```python
from pathlib import Path
import numpy as np
import glob
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import shutil
```

## Declaração de Constantes e Modelos

```python
model = load_model('image_quality (1).h5', compile=False)
```

## Funções Necessárias

```python
def load_and_prepare_image(img_path, target_size=(224, 224)):
    """
    Load and prepare an image for classification.

    Parameters:
    - img_path: string containing the path of the image to be loaded.
    - target_size: tuple (width, height) representing the target size of the image.

    Returns:
    - img_array: ndarray containing the loaded and prepared image.
    """
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img).astype('uint8')
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0 
    return img_array


def classify_image(model, img_array):
    """
    Classify an image using a pre-trained classification model.

    Parameters:
    - model: pre-trained classification model.
    - img_array: ndarray containing the image to be classified.

    Returns:
    - result: string indicating the classification of the image ("bad" or "good").
    """
    predictions = model.predict(img_array)
    result = np.argmax(predictions, axis=1)
    if result[0] == 0:
        return "bad"
    else:
        return "good"
```

## Inferência do modelo em um conjunto de imagens

```python
for img_path in glob.glob('Lote 1 e 2 - Riser/*.jpg'):
    img_array = load_and_prepare_image(img_path)
    predicted_class = classify_image(model, img_array)
    shutil.copy(img_path, 'Lote 1 e 2 - Riser - Qualidade/' + predicted_class + '/' + img_path.split('/')[-1])
```
