## Exemplos disponíveis:

- Versões locais:
    - dam_video_with_sam: usa o SAM e DAM em um vídeo. Segmenta objetos em um conjunto de frames de um vídeo e, em seguida, gera descrições detalhadas das regiões segmentadas.
    - dam_video_with_sam2: usa o SAM2 e DAM  em um vídeo. Ele realiza segmentação com base em pontos ou caixas fornecidos no primeiro frame do vídeo e propaga essa segmentação para os frames subsequentes. 
    - dam_with_sam_self_contained.py: versão auto-suficiente para realizar SAM e DAM de uma imagem localmente. Ele não requer a instalação de pacotes externos como dam, pois tudo que é necessário está incluído diretamente no script ou é gerenciado por bibliotecas como transformers e torch.
    - dam_with_sam: SAM + DAM para descrever uma região mascarada em uma imagem (totalmente local).
- Versões com API para LLM:
    - query_dam_server_raw: interagir com a API da OpenAI para descrever uma região mascarada em uma imagem (com a biblioteca requests).
    - query_dam_server_video: interagir com a API da OpenAI para descrever um objeto em uma região mascarada ao longo de um vídeo.
    - query_dam_server: interagir com a API da OpenAI para descrever uma região mascarada em uma imagem (com a biblioteca openai).
