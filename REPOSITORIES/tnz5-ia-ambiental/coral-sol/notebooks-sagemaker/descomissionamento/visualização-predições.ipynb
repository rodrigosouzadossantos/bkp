## Objetivo: Analisar os casos de alta probabilidade, para entender melhor os resultados do modelo.

## Importações necessárias

```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

## Visualização dos eventos de cruzamento

```python
caminho_csv = 'eventos.csv'
caminho_videos = "../Videos-Teste/downloads"

df = pd.read_csv(caminho_csv)
df = df[df['Cruzamento'] > 0.8]
videos = sorted(df['Video'].unique())

for video in videos:
    df_video = df[df['Video'] == video].copy()
    df_video['Minuto'] = df_video['Tempo'].apply(lambda x: int(x.split(':')[0]) if ':' in x else -1)
    df_video['Segundo'] = df_video['Tempo'].apply(lambda x: int(x.split(':')[1]) if ':' in x else -1)
    df_video = df_video.sort_values(['Minuto', 'Segundo'])
    exemplos = df_video.head(5)
    video_path = f"{caminho_videos}/{video}.mp4"
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Erro ao abrir o vídeo: {video}")
        continue
    fps = cap.get(cv2.CAP_PROP_FPS)
    ultimo_minuto = -1
    for _, row in exemplos.iterrows():
        tempo = row['Tempo']
        cruzamento = row['Cruzamento']
        if ':' not in tempo:
            continue
        minutos, segundos = map(int, tempo.split(':'))
        if minutos == ultimo_minuto:
            continue
        ultimo_minuto = minutos
        frame_number = int((minutos * 60 + segundos) * fps)
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        ret, frame = cap.read()
        if ret:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(10, 6))
            plt.imshow(frame_rgb)
            plt.title(f"Vídeo: {video}\nTempo: {tempo} | Cruzamento: {cruzamento:.2f}")
            plt.axis('off')
            plt.show()
    cap.release()
```
