## Importações Necessárias

```python
!pip install shapely
import os
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
from shapely.geometry import Polygon, MultiPolygon
```

## Funções Necessárias

```python
def plot_image_with_annotations(image_path: str, annotation_path: str) -> None:
    """
    Plot an image with its annotations.

    Args:
        image_path (str): The path to the image file.
        annotation_path (str): The path to the annotation file.
    """
    try:
        # Load the image
        image = Image.open(image_path)
        fig, ax = plt.subplots(1)
        ax.imshow(image)

        # Load annotations
        annotations = load_annotations(annotation_path)
        if annotations is None:
            print(f"No annotations found for {image_path}")
            return

        # Add annotations to the image
        for annotation in annotations['shapes']:
            points = annotation['points']
            polygon = Polygon(points)
            if isinstance(polygon, Polygon):
                patch = patches.Polygon(np.array(polygon.exterior.coords), closed=True, edgecolor='r', facecolor='none')
                ax.add_patch(patch)
            elif isinstance(polygon, MultiPolygon):
                for poly in polygon:
                    patch = patches.Polygon(np.array(poly.exterior.coords), closed=True, edgecolor='r', facecolor='none')
                    ax.add_patch(patch)

        plt.show()

    except Exception as e:
        print(f"Error processing annotation for {image_path}: {e}")

def load_annotations(annotation_path: str) -> dict:
    """
    Load annotations from a JSON file.

    Args:
        annotation_path (str): The path to the annotation file.

    Returns:
        dict: The annotations.
    """
    try:
        with open(annotation_path, 'r') as file:
            annotations = json.load(file)
        return annotations
    except Exception as e:
        print(f"Error reading annotation file {annotation_path}: {e}")
        return None


def process_dataset_images(dataset_folder: str) -> None:
    """
    Process images in the dataset folder and plot them with annotations.

    Args:
        dataset_folder (str): The path to the dataset folder.
    """
    for image_filename in os.listdir(dataset_folder):
        try:
            if is_image_file(image_filename):
                print(image_filename)
                image_path = os.path.join(dataset_folder, image_filename)
                annotation_path = get_annotation_path(dataset_folder, image_filename)

                if os.path.exists(annotation_path):
                    plot_image_with_annotations(image_path, annotation_path)
                else:
                    print(f"No corresponding JSON file for {image_filename}")
        except Exception as e:
            print(f"Error processing file {image_filename}: {e}")


def is_image_file(filename: str) -> bool:
    """
    Check if a file is an image based on its extension.

    Args:
        filename (str): The name of the file.

    Returns:
        bool: True if the file is an image, False otherwise.
    """
    return filename.lower().endswith(('.jpg', '.jpeg', '.png'))


def get_annotation_path(dataset_folder: str, image_filename: str) -> str:
    """
    Get the corresponding annotation file path for a given image file.

    Args:
        dataset_folder (str): The path to the dataset folder.
        image_filename (str): The name of the image file.

    Returns:
        str: The path to the corresponding annotation file.
    """
    base_filename = os.path.splitext(image_filename)[0]
    return os.path.join(dataset_folder, f"{base_filename}.json")

```

## Visualização das imagens do dataset

```python
dataset_folder = 'Dataset/PUC - Coral - Segmentado/PUC - Coral - F'
process_dataset_images(dataset_folder)
```
