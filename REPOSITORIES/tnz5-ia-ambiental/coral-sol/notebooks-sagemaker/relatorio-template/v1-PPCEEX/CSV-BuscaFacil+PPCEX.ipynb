Objetivo: Este notebook cruza os dados de duas planilhas (PPCEX e Busca-Facil), filtrando registros válidos com base em colunas específicas, para gerar um novo arquivo CSV contendo informações selecionadas e consolidadas sobre ordens de serviço, identificadores, caminhos completos, códigos de relatório e tamanhos, removendo duplicidades e valores inválidos.

## Importações necessárias

```python
!pip install openpyxl
import pandas as pd
```

## Declaração de Constantes e Modelos

```python
local_csv = 'PPCEX.csv'
local_xlsx = 'Busca-Facil.xlsx'
```

## Geração da planilha os_caminho_completo

```python
local_df = pd.read_csv(local_csv, delimiter=",")
local_df = local_df.dropna(subset=['OS'])
xlsx_df = pd.read_excel(local_xlsx, engine='openpyxl', header=0)

local_df['OS'] = local_df['OS'].astype(str).str.replace(r'\.0+$', '', regex=True)
xlsx_df['Unnamed: 4'] = xlsx_df['Unnamed: 4'].astype(str).str.replace(r'\.0+$', '', regex=True)

merged_df = pd.merge(local_df, xlsx_df, left_on='OS', right_on='Unnamed: 4')
merged_df = merged_df[~merged_df['Unnamed: 13'].astype(str).str.strip().isin(['0', '0B', '0.0', '0.0B'])]

result_df = merged_df[['OS', 'Atualizado em 2025-08-10', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 13']]
result_df = result_df.dropna(subset=['Unnamed: 3'])
result_df = result_df.drop_duplicates()
result_df.columns = ['OS', 'ID', 'Caminho_Completo', 'Codigo_Relatorio', 'Tamanho']
result_df.to_csv('os_caminho_completo.csv', index=False)
```
