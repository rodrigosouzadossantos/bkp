## Orquestrador - Notebook Template - Coral-Sol

Orquestrador para a geração de relatórios baseados em inteligência artificial, com foco em análises sobre o coral-sol em uma operação. A solução busca otimizar o processo de organização e apresentação dos dados, garantindo maior eficiência e precisão.

## Etapa 1 - Geração da tabela inicial: execução do script generate_data.py, responsável por gerar um arquivo .csv e algumas pastas, contendo todos os dados necessários para o relatório.

Primeiramente, realizamos a verificação do FPS (Frames por Segundo) do vídeo inicial, com o objetivo de determinar a quantidade de frames a serem ignorados na geração dos dados. Caso a intenção seja processar 1 frame por segundo, recomenda-se utilizar o primeiro valor obtido. Alternativamente, para um intervalo maior, como 1 frame a cada 3 segundos, deve-se optar pelo segundo valor.

````python
!python3 verify_fps.py "Jaqueta/videos"
````

Parâmetros de entrada:
```
--operacao: Diretório contendo os vídeos que serão processados. O caminho especificado deve incluir os arquivos de vídeo nos formatos suportados (.mp4, .avi, .mkv).

--out_frames: Diretório onde os frames extraídos dos vídeos serão salvos. Cada frame será armazenado no formato .npy.

--out_masks: Diretório para salvar as máscaras resultantes da segmentação realizada pelo modelo YOLO. As máscaras também serão armazenadas no formato .npy.

--out_scores: Diretório para salvar os scores associados às máscaras geradas pelo modelo.

--out_csv: Caminho do arquivo CSV onde os resultados da inferência serão registrados. Este arquivo conterá informações como timestamp, coordenadas extraídas e scores.

--model_path: Caminho do arquivo que contém o modelo YOLO utilizado para inferência. Certifique-se de fornecer o arquivo correto (.pt) compatível com a tarefa de segmentação desejada.

--frame_skip: Número de frames a serem ignorados entre cada inferência. Por exemplo, se o valor for 45, o script processará 1 frame a cada 45 frames.

--x, --y: Coordenadas iniciais (x, y) do ROI (Região de Interesse) dentro do frame. Esta é a área específica onde os dados serão extraídos.

--w, --h: Largura (w) e altura (h) do ROI. Define o tamanho da área que será analisada dentro do frame.
```


````python
!python3 generate_data.py \
--operacao "Jaqueta/videos" \
--out_frames "Jaqueta/frames" \
--out_masks "Jaqueta/masks" \
--out_scores "Jaqueta/scores" \
--out_csv "Jaqueta/resultados_inferencia_1.csv" \
--model_path "coral_yolov11_segm_2k_V6.4_parcial.pt" \
--frame_skip 45 \
--x 706 \
--y 14 \
--w 123 \
--h 67
````

## Etapa 2 - Execução de scripts complementares

O script generate_plots.py gera gráficos a partir dos resultados de inferência registrados em um arquivo CSV.

Parâmetros de Entrada:
```
--csv_path: Caminho para o arquivo CSV contendo os resultados da inferência. Este arquivo deve incluir informações como paths dos frames e máscaras, scores, timestamps e nomes dos vídeos que foram processados.

--output_dir: Diretório onde os gráficos gerados serão salvos. Os gráficos serão armazenados no formato .png, com cada arquivo representando um trecho analisado do vídeo.
```


````python
!python3 generate_plots.py \
--csv_path "Jaqueta/resultados_inferencia_1.csv" \
--output_dir "Jaqueta/plots_from_csv"
````

O script generate_maps.py gera um mapa visual da operação com base nos dados de coordenadas (North e East) registrados em um arquivo CSV. 

Parâmetros de Entrada:
```  
--csv_path: Caminho para o arquivo CSV contendo os resultados da inferência. Este arquivo deve incluir informações como coordenadas North e East, além dos scores relacionados à presença de coral.

--output_file: Caminho e nome do arquivo de saída onde o mapa será salvo. O mapa será gerado no formato .png.
```


````python
!python3 generate_maps.py \
--csv_path "Jaqueta/resultados_inferencia_1.csv" \
--output_file "Jaqueta/mapa_operacao.png"
````

O script generate_serie.py gera gráficos de confiança a partir dos resultados registrados em um arquivo CSV, com base nos scores inferidos em vídeos.

Parâmetros de Entrada:

```
--csv_path: Caminho para o arquivo CSV contendo os resultados da inferência. Este arquivo deve incluir informações como scores e timestamps de cada vídeo processado.

--threshold: Valor limite de confiança utilizado como linha de corte nos gráficos. Scores acima desse valor serão considerados significativos.

--space: Número de espaços em branco a serem inseridos entre os dados de diferentes vídeos nos gráficos, para facilitar a separação visual entre os vídeos.

--label_interval: Intervalo de tempo (em número de frames) para exibir os rótulos de timestamp no eixo X dos gráficos.

--ma_window: Tamanho da janela utilizado para calcular a média móvel dos scores de confiança. Esse valor suaviza os dados para uma melhor visualização.

--output_dir: Diretório onde os gráficos gerados serão salvos. Dois gráficos serão criados: um exibindo os scores de confiança e outro com os scores suavizados pela média móvel.
```

````python
!python3 generate_serie.py \
--csv_path "Jaqueta/resultados_inferencia_1.csv" \
--threshold 0.5 \
--space 5 \
--label_interval 30 \
--ma_window 5 \
--output_dir "Jaqueta"
````

O script generate_table.py gera uma tabela formatada a partir dos dados de inferência registrados em um arquivo CSV, filtrando outliers das coordenadas e organizando as informações relevantes em colunas específicas.

Parâmetros de Entrada:

```
--csv_path: Caminho para o arquivo CSV contendo os resultados da inferência. Este arquivo deve incluir informações como coordenadas (north, east), scores (scores_list), nomes de vídeos e timestamps.

--output_file: Caminho e nome do arquivo de saída onde a tabela formatada será salva. A saída será gerada no formato .csv.
```

````python
!python3 generate_table.py \
--csv_path "Jaqueta/resultados_inferencia_1.csv" \
--output_file "Jaqueta/tabela_formatada.csv"
````

## Consolidação dos resultados: Com os resultados gerados pelos quatro scripts, todos os dados e imagens são agrupados e organizados. Por fim, é criado um arquivo .html contendo o relatório completo, apresentando as informações de forma visualmente clara e estruturada.

O script generate_html.py gera um relatório em formato HTML com base nos resultados de inferência e análises realizadas. Este relatório inclui informações da operação, gráficos e mapas gerados, além de uma tabela detalhada com os eventos identificados.

Parâmetros de Entrada:

```
--plots_dir: Diretório contendo as imagens dos gráficos gerados durante a análise. Estas imagens serão apresentadas no relatório.

--csv_path: Caminho para o arquivo CSV contendo os resultados formatados da análise, como informações sobre os eventos identificados.

--output_html: Nome do arquivo HTML que será gerado como saída. Este arquivo será salvo no diretório especificado no parâmetro --os_number.

--description: Descrição da operação realizada. Este texto será exibido no relatório.

--os_number: Número da OS (Ordem de Serviço) da operação. Será utilizado para identificar a operação no relatório.

--model_name: Nome do modelo de IA utilizado para análise. Será exibido na seção de informações do modelo.

--model_desc: Descrição do modelo de IA, incluindo suas recomendações e limitações.

--model_time: Tempo total de processamento realizado pelo modelo. Este valor será exibido no relatório.

--model_tax: Taxa de quadros analisada pelo modelo (e.g., 1 frame a cada 3 segundos).

--conf_img: Caminho para o gráfico de confiança do modelo de IA ao longo do tempo.

--mm_img: Caminho para o gráfico da média móvel dos scores de confiança.

--map_img: Caminho para o mapa gerado com as coordenadas da operação.
```


````python
!python3 generate_html.py \
--plots_dir "Jaqueta/plots_from_csv" \
--csv_path "Jaqueta/tabela_formatada.csv" \
--output_html relatorio.html \
--description "Operação com coral-sol." \
--os_number "Jaqueta" \
--model_name "Modelo coral_yolov11_segm_2k_V6.4_parcial" \
--model_desc "Modelo de segmentação para detecção de coral-sol." \
--model_time "45 minutos e 19 segundos" \
--model_tax "1 frames a cada 3 segundo" \
--conf_img "Jaqueta/grafico_confiança.png" \
--mm_img "Jaqueta/grafico_media_movel.png"\
--map_img "Jaqueta/mapa_operacao.png"
````
