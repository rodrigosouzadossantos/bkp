# **Clusterizer Evaluation**

A tabela que você mostrou é o resultado do módulo:

**“Automatic Clusterizer Evaluation (KMeans × HDBSCAN)”**

Ele compara os dois algoritmos e decide qual representou melhor a estrutura real do dataset.

Aqui está a explicação linha por linha:

---

# **1) KMeans Score = 1**

Isso significa:

- O KMeans venceu **apenas 1** dos critérios avaliados  
- Em quase todos os outros pontos, teve desempenho inferior ao HDBSCAN  

Isso normalmente acontece quando:

- Os dados **não são bem separados de forma esférica**  
- O dataset tem **subgrupos densos**  
- Existem **muitos padrões diferentes**  
- O dataset é **complexo ou irregular**  

KMeans funciona melhor em dados “organizados” — mas parece não ser o caso aqui.

---

# **2) HDBSCAN Score = 4**

O HDBSCAN venceu **4 dos critérios avaliados**, o que indica que:

✔ Detectou clusters reais de forma mais natural  
✔ Teve melhor “ajuste” à estrutura do dataset  
✔ Lidou melhor com regiões densas / irregulares  
✔ Foi mais robusto a ruído  

Quando HDBSCAN ganha com diferença grande assim, é um forte sinal de que:

**o dataset tem densidades diferentes, clusters pequenos e zonas ruidosas.**

Isso é comum em:

- conjuntos de imagens submarinas  
- cenários naturais  
- ambientes com iluminação variável  
- imagens com objetos esparsos  

---

# **3) Winner = HDBSCAN**

A conclusão automática do sistema:

**HDBSCAN representa melhor a estrutura real do dataset.**

O HDBSCAN é superior quando:

- clusters têm tamanhos muito diferentes  
- existem “ilhas” de dados densos  
- há bastante ruído ou amostras difíceis  
- geometria dos clusters não é esférica  
- existe grande diversidade visual  

Provavelmente essas características **existem no seu dataset**.

---

# **4) Noise Ratio = 232**

Esse número significa:

**232 imagens foram classificadas como “noise” pelo HDBSCAN**  
(ou seja, receberam label -1).

Interpretação:

- Essas imagens **não pertencem a nenhum cluster denso**  
- São “casos isolados” ou “diferentes de todo o resto”  
- Possivelmente são:  
  • imagens muito escuras  
  • muito claras  
  • borradas  
  • ângulos diferentes  
  • presença de objetos raros  
  • ruído visual  
  • imagens corrompidas  
  • cenários totalmente diferentes  

Um número alto de noise indica um dataset **bem heterogêneo**.

---

# **5) Alignment = 0.000**

O alinhamento mede:

**“Os clusters do KMeans e do HDBSCAN concordam?”**

Ele calcula a força da diagonal no heatmap KMeans × HDBSCAN.

Valor 0.000 significa:

**Os dois algoritmos reconheceram estruturas muito diferentes.**  
**Quase não existe correspondência entre os clusters dos dois métodos.**

Implica:

- KMeans está forçando agrupamentos
- HDBSCAN está encontrando agrupamentos naturais completamente diferentes
- O dataset não é adequado para KMeans
- A clusterização real é “guiada pela densidade”, não por posição esférica

Em resumo:

> Os dois algoritmos não enxergam o dataset da mesma forma, e  
> o HDBSCAN claramente faz mais sentido para a estrutura visual real.

---

# **Conclusão Geral**

Com base nos valores:

- **HDBSCAN é MUITO melhor para este dataset**  
- KMeans falha em capturar a estrutura  
- O dataset contém muita variedade visual  
- Há muitos outliers ou casos raros  
- Os padrões naturais estão mais ligados à densidade  
- O dataset não forma clusters “redondos” → ruim para KMeans  
- HDBSCAN detecta padrões mais sofisticados e formas não-lineares  

---

# **O que fazer com isso na prática**

## **1. Use HDBSCAN como clusterizador principal**
Ele reflete melhor a realidade do dataset.

## **2. Analise as imagens marcadas como noise (-1)**
Elas são:
- casos extremos  
- divergentes  
- potencialmente problemáticas  
- importantes para entender falhas de modelos ML  

## **3. Use o heatmap para entender diferenças profundas**
KMeans provavelmente distribuiu imagens de forma arbitrária.

## **4. Reforce validações visuais**
Abra os clusters do HDBSCAN para identificar:
- grupos temáticos  
- subcenários  
- padrões reais  

## **5. Considerar limpeza do dataset**
Se muitos noise forem irrelevantes → podem ser removidos.
