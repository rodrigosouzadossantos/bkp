```python
import unicodedata
import re
def normalize(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = unicodedata.normalize("NFD", text)
    text = "".join(c for c in text if unicodedata.category(c) != "Mn")
    text = re.sub(r"[^a-z0-9]+", " ", text)
    return text.strip()
def only_digits(text):
    return "".join(c for c in text if c.isdigit())

def clean_path_folders(path,values):
    if not isinstance(path,str):
        return path
    folders=[f for f in path.split("/") if f]
    norm_vals=[normalize(v) for v in values if v]
    word_sets=[set(v.split()) for v in norm_vals]
    os_digits=""
    for v in values:
        if isinstance(v,str) and v.isdigit():
            os_digits=v
            break
    cleaned=[]
    for folder in folders:
        nf=normalize(folder)
        nf_words=set(nf.split())
        remove=False
        for vw in word_sets:
            if not vw:
                continue
            common=len(nf_words & vw)
            if common>=max(1,int(len(vw)*0.5)):
                remove=True
                break
        if (not remove) and os_digits:
            f_digits="".join(c for c in folder if c.isdigit())
            if os_digits in f_digits:
                remove=True
        if not remove:
            cleaned.append(folder)
    return "/".join(cleaned)+"/"

cols=["os","desc","company","vessel",'bucket']

p = df.filter(
    pl.col("path").str.contains(r"/\d{4}/")
).select( [ 'os', 'desc', 'company', 'vessel','bucket', 'path' ] ).with_columns([
    pl.struct(["path"]+cols).map_elements(
        lambda r: clean_path_folders(
            r["path"],[r[c] for c in cols]
        )
    ).alias("path_clean")
])

# .with_columns([
#     pl.col( 'path' )
#         .str.to_lowercase( )
#         .str.strip_chars( )
#         .str.normalize( 'NFKD' )
#         .str.replace_all( r'[\u0300-\u036f]', '' )

# ])
    # pl.struct([ 'os', 'desc', 'company', 'vessel', 'path' ]).map_elements(
    #     lambda row: (
    #         row["path"]
    #         .replace(row["os"] or "", "")
    #         .replace(row["desc"] or "", "")
    #     )
    #   ).alias("path_clean")

      #.str.replace(r"^.*/(\d{4})/", r"$1-").alias('tst')
      # .str.replace( pl.col( 'desc' ), '', literal = False )
      # .str.replace( pl.col( 'os' ), '', literal = False )
# ])
```

```python
from keybert import KeyBERT

kw_model = KeyBERT(model="all-mpnet-base-v2")

def label_with_keybert(df):
    labs=[]
    clusters=df["class_cluster"].unique()
    for cl in clusters:
        sub=df.filter(pl.col("class_cluster")==cl)
        txt=sub["class_norm"].to_list()
        doc=" ".join(txt)
        kws=kw_model.extract_keywords(
            doc,
            keyphrase_ngram_range=(1,3),
            stop_words=None,
            use_mmr=True,
            diversity=0.4,
            top_n=3
        )
        ph=[k for k,_ in kws]
        ph=[p.capitalize() for p in ph]
        if len(ph)==0:
            lab="Desconhecido"
        elif len(ph)==1:
            lab=ph[0]
        else:
            lab=f"{ph[0]} ({' / '.join(ph[1:])})"
        labs.append({"class_cluster":cl,
                     "cluster_label":lab})
    mapdf=pl.DataFrame(labs)
    return df.join(mapdf,on="class_cluster",
                   how="left")


import numpy as np
from sklearn.feature_extraction.text import \
    CountVectorizer
from sentence_transformers import \
    SentenceTransformer

st_model=SentenceTransformer(
    "all-mpnet-base-v2"
)

def label_with_centroid(df):
    labs=[]
    clusters=df["class_cluster"].unique()
    for cl in clusters:
        sub=df.filter(pl.col("class_cluster")==cl)
        txt=sub["class_norm"].to_list()
        emb=st_model.encode(txt)
        cen=np.mean(emb,axis=0)
        vec=CountVectorizer(
            ngram_range=(1,3),
            stop_words=None,
            min_df=1,
            token_pattern=r"(?u)\\b\\w+\\b"
        )
        vec.fit(txt)
        cand=vec.get_feature_names_out()
        cemb=st_model.encode(cand)
        num=np.dot(cemb,cen)
        den=np.linalg.norm(cemb,axis=1)*\
            np.linalg.norm(cen)
        sim=num/den
        idx=sim.argsort()[::-1][:3]
        ph=[cand[i].capitalize() for i in idx]
        if len(ph)==0:
            lab="Desconhecido"
        elif len(ph)==1:
            lab=ph[0]
        else:
            lab=f"{ph[0]} ({' / '.join(ph[1:])})"
        labs.append({"class_cluster":cl,
                     "cluster_label":lab})
    mapdf=pl.DataFrame(labs)
    return df.join(mapdf,on="class_cluster",
                   how="left")
```

```python
import polars as pl
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer

st_model = SentenceTransformer("all-mpnet-base-v2")

def centroid_label(texts, top_n=3):
    # 1. basic cleanup
    texts = [t for t in texts if t and isinstance(t, str)]
    if len(texts) == 0:
        return "Desconhecido"
    # 2. embed cluster texts
    emb = st_model.encode(texts)
    cen = np.mean(emb, axis=0)
    # --------------------------------------
    # 3. extract candidate phrases robustly
    # --------------------------------------
    vec = CountVectorizer(
        ngram_range=(1,3),
        stop_words=None,              # keep Portuguese
        min_df=1,
        token_pattern=r"(?u)\b\w+\b"  # capture codes, pidf-3 тЖТ pidf, 3
    )
    try:
        vec.fit(texts)
    except ValueError:
        return "Desconhecido"
    cand = vec.get_feature_names_out()
    if len(cand) == 0:
        return "Desconhecido"
    # 4. embed candidates
    cemb = st_model.encode(cand)
    # 5. cosine similarity with centroid
    num = np.dot(cemb, cen)
    den = np.linalg.norm(cemb, axis=1) * np.linalg.norm(cen)
    sim = num / den
    # 6. pick best N
    idx = sim.argsort()[::-1][:top_n]
    ph = [cand[i].capitalize() for i in idx]
    if len(ph) == 1:
        return ph[0]
    return f"{ph[0]} ({' / '.join(ph[1:])})"


def label_with_centroid(df):
    labels = []
    for cl in df["class_cluster"].unique():
        sub = df.filter(pl.col("class_cluster") == cl)
        texts = sub["class_norm"].to_list()
        label = centroid_label(texts)
        labels.append({"class_cluster": cl,
                       "cluster_label": label})
    mapdf = pl.DataFrame(labels)
    return df.join(mapdf, on="class_cluster", how="left")
```

```python
# pl.col( 'class_norm').is_null() == 3_261
def human_bytes( col ) :
  return (
    pl.when( col < 1024 ).then( col.cast(pl.Utf8) + " B" )
      .when( col < 1024**2 ).then( ( col / 1024 ).round( 2 ).cast( pl.Utf8 ) + " KB" )
      .when( col < 1024**3 ).then( ( col / 1024**2 ).round( 2 ).cast( pl.Utf8 ) + " MB" )
      .when( col < 1024**4 ).then( ( col / 1024**3 ).round( 2 ).cast( pl.Utf8 ) + " GB" )
      .when( col < 1024**5 ).then( ( col / 1024**4 ).round( 2 ).cast( pl.Utf8 ) + " TB" )
      .when( col < 1024**6 ).then( ( col / 1024**5 ).round( 2 ).cast( pl.Utf8 ) + " PB" )
      .when( col < 1024**7 ).then( ( col / 1024**6 ).round( 2 ).cast( pl.Utf8 ) + " EB" )
      .when( col < 1024**8 ).then( ( col / 1024**7 ).round( 2 ).cast( pl.Utf8 ) + " ZB" )
      .otherwise( ( col / 1024**8 ).round( 2 ).cast( pl.Utf8 ) + " YB" )
      
)

import plotly.graph_objects as go
import polars as pl

def build_sankey(df: pl.DataFrame):
    # -----------------------------------------------------
    # Ensure required columns exist
    # -----------------------------------------------------
    for col in ["class_norm", "company_norm", "bucket"]:
        if col not in df.columns:
            raise ValueError(f"Column '{col}' missing in dataframe.")
    # Remove nulls
    dfc = df.filter(
        pl.col("class_norm").is_not_null() &
        pl.col("company_norm").is_not_null() &
        pl.col("bucket").is_not_null()
    )
    df_with_cluster = dfc.with_columns(
        pl.col("class_norm").map_elements(
            lambda x: suit_to_cluster.get(x, -1),
            return_dtype=pl.Int32
        ).alias("class_cluster")
    )
    valid_clusters = (
        df_with_cluster
        .group_by("class_cluster")
        .agg(pl.len().alias("count"))
        .filter(pl.col("count") > 160)
        .select("class_cluster")
        .to_series()
    )
    valid_clusters_df = pl.DataFrame({"class_cluster": valid_clusters.to_list()})
    dfc = df_with_cluster.join(valid_clusters_df, on="class_cluster", how="inner")
    # -----------------------------------------------------
    # Build unique lists of nodes
    # -----------------------------------------------------
    class_list   = dfc["class_norm"].unique().to_list()
    company_list = dfc["company_norm"].unique().to_list()
    bucket_list  = dfc["bucket"].unique().to_list()
    # Concatenate into a single node list
    nodes = class_list + company_list + bucket_list
    # map each node to an index
    node_index = {name: i for i, name in enumerate(nodes)}
    # -----------------------------------------------------
    # Build links: CLASS тЖТ COMPANY
    # -----------------------------------------------------
    df_class_company = (
        dfc
        .group_by(["class_norm", "company_norm"])
        .agg(pl.len().alias("count"))
    )
    sources_1 = [
        node_index[row["class_norm"]] for row in df_class_company.iter_rows(named=True)
    ]
    targets_1 = [
        node_index[row["company_norm"]] for row in df_class_company.iter_rows(named=True)
    ]
    values_1 = [
        row["count"] for row in df_class_company.iter_rows(named=True)
    ]
    # -----------------------------------------------------
    # Build links: COMPANY тЖТ BUCKET
    # -----------------------------------------------------
    df_company_bucket = (
        dfc
        .group_by(["company_norm", "bucket"])
        .agg(pl.len().alias("count"))
    )
    sources_2 = [
        node_index[row["company_norm"]] for row in df_company_bucket.iter_rows(named=True)
    ]
    targets_2 = [
        node_index[row["bucket"]] for row in df_company_bucket.iter_rows(named=True)
    ]
    values_2 = [
        row["count"] for row in df_company_bucket.iter_rows(named=True)
    ]
    # -----------------------------------------------------
    # Build Sankey Graph
    # -----------------------------------------------------
    link = dict(
        source=sources_1 + sources_2,
        target=targets_1 + targets_2,
        value=values_1 + values_2,
        color=["rgba(100,149,237,0.4)"] * len(sources_1) +
              ["rgba(60,179,113,0.4)"] * len(sources_2)
    )
    node = dict(
        label=nodes,
        pad=20,
        thickness=20,
        color="rgba(50, 50, 50, 0.8)"
    )
    fig = go.Figure(data=[go.Sankey(node=node, link=link)])
    fig.update_layout(
        title="Service Class тЖТ Company тЖТ Bucket Flow",
        font_size=12,
        height=700
    )
    return fig
# ( df.filter( pl.col( 'class_norm').is_not_null() )
#     .with_columns(
#       pl.col("class_norm").map_elements(
#             lambda x: suit_to_cluster.get(x, -1),
#             return_dtype=pl.Int32
#         ).alias("class_cluster")
#     ).select( [ 'class_norm', "class_cluster" ] )
#     .group_by("class_norm")
#     .agg( pl.len() ).sort( 'len', descending = True )
#     .filter(
#         pl.col( 'len' ) > 160
#     )
#     # .filter(
#     #     pl.col("class_norm")
#     #         .cast( pl.String )
#     #         .str.contains("pidf", literal=True )
#     # )
# ).select( 'class_norm' )
# # ).select( 'len' ).sum( )
# # ).describe( percentiles = [0.05,0.2,0.35,0.5,0.65,0.8,0.95] )
cluster_class = (
    df
    .filter(pl.col("class_norm").is_not_null())
    .with_columns(
        pl.col("class_norm").map_elements(
            lambda x: suit_to_cluster.get(x, -1),
            return_dtype=pl.Int32
        ).alias("class_cluster")
    )
    .group_by("class_cluster")
    .agg(pl.len().alias("count"))
    .filter(pl.col("count") > 160)     # same threshold as your example
    # .select("class_cluster")
    # .to_series()
    # .to_list()
)
#cluster_class.filter(pl.col('class_cluster')!=0).group_by('class_cluster').agg([
#    human_bytes(pl.col('files_size').sum())
#])



# main_classes
```

```python
def human_bytes_py(x):
    if x < 1024:
        return f"{x} B"
    elif x < 1024**2:
        return f"{x/1024:.2f} KB"
    elif x < 1024**3:
        return f"{x/1024**2:.2f} MB"
    elif x < 1024**4:
        return f"{x/1024**3:.2f} GB"
    else:
        return f"{x/1024**4:.2f} TB"

df.select("files_size").describe().with_columns(
    pl.col("files_size")
      .map_elements(lambda x: human_bytes_py(int(x)),
                    return_dtype=pl.String)
)
```
