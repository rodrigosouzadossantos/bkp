```python
import unicodedata,re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

custom_stopwords = set( STOPWORDS )
custom_stopwords.update( [ ] )

def normalize_word(t):
    t=t.lower()
    t=unicodedata.normalize("NFD",t)
    t="".join(c for c in t if unicodedata.category(c)!="Mn")
    t=re.sub(r"[^a-z0-9 ]+"," ",t)
    return t.strip()

def _build_wordcloud(df):
    words=[]
    for s in df["class_norm"].to_list():
        if isinstance(s,str):
            txt=normalize_word(s)
            for w in txt.split():
                # if w and (w not in pt_stopwords):
                    words.append(w)
    text=" ".join(words)
    wc=WordCloud(
      width = 500,
      height=350,
      background_color="white",
      colormap="viridis",
      max_words=200,
      stopwords=custom_stopwords,
      min_font_size=10
    ).generate(text)
    
    fig,ax=plt.subplots(figsize=(14,7))
    ax.imshow(wc,interpolation="bilinear")
    ax.axis("off")
    return fig
```

```python
import unicodedata,re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

import numpy as np
def build__wordcloud(df):
    x, y = np.ogrid[:300, :300]
    mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
    mask = 255 * mask.astype(int)
    
    custom_stopwords = set( STOPWORDS )
    custom_stopwords.update( [ 'e', 'em', 'ou', 'de' ] )

    words = [ ]
    for s in df[ 'class_norm' ].to_list( ) :
      [ words.append( w )
          for w in s.split( )
          # if isinstance( s, str )
      ]

    text=" ".join(words)
    wc=WordCloud(
      width = 500,
      height=350,
      background_color="white",
      colormap="viridis",
      max_words=100,
      stopwords=custom_stopwords,
      min_font_size=10,
      mask = mask
    ).generate(text)
    
    fig, ax = plt.subplots( figsize=( 14, 7 ) )
    ax.imshow( wc, interpolation = 'bilinear' )
    ax.axis( 'off' )
    return fig
```
