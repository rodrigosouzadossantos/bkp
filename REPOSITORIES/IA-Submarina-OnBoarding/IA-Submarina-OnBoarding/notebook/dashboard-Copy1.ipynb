```python
import ipywidgets as widgets
import plotly.express as px
from IPython.display import display

# dataset de exemplo
df = px.data.gapminder()

# widgets de filtro
ano = widgets.IntSlider(value=2007, min=1952, max=2007, step=5, description="Ano:")

continente = widgets.Dropdown(options=df.continent.unique(), description="Continente:")
# function de atualização
def atualizar_dashboard(ano, continente):
    filtrado = df[(df.year == ano) & (df.continent == continente)]
    fig = px.scatter(filtrado, x="gdpPercap", y="lifeExp", size="pop", color="country")
    fig.show()
# cria painel
ui = widgets.VBox([ano, continente])
out = widgets.interactive_output(atualizar_dashboard, {"ano": ano, "continente": continente})
display(ui, out)
```

```python

# ===========================================================
# KPI cards
# ===========================================================
def create_card(title, value):
    style = """
        padding: 20px;
        border-radius: 10px;
        background-color: #1f2937;
        color: white;
        width: 300px;
        text-align: center;
        box-shadow: 0 4px 10px rgba(0,0,0,0.3);
        font-family: Arial, sans-serif;
    """
    return widgets.HTML(
        value=f"""
        <div style="{style}">
            <div style="font-size: 18px; opacity: 0.8;">{title}</div>
            <div style="font-size: 42px; font-weight: bold; margin-top: 10px;">
                {value}
            </div>
        </div>
        """
    )


# Formatting function
def format_tb( value ):
  return (
    "{} TB".format( round( ( value / 1024**5 ), 2 ) )
  )

cards_row = widgets.HBox(
    [
        create_card("Total Storage", format_tb( df.select( ["files_size"] ).sum().item() ) ),
        widgets.HTML("<div style='width:20px;'></div>"),
        create_card("Ingestion Rate", "540 MB/s")
    ],
    layout=widgets.Layout(justify_content="flex-start")
)

# ===========================================================
# Slider (year selector)
# ===========================================================
df_year_bounds = (
    df
    .with_columns(pl.col("files_oldest").alias("date"))
    .select([
        pl.col("date").dt.year().min().alias("min_year"),
        pl.col("date").dt.year().max().alias("max_year")
    ])
)
min_year = df_year_bounds["min_year"][0]
max_year = df_year_bounds["max_year"][0]
slider_year = widgets.IntSlider(
    value=max_year,
    min=min_year,
    max=max_year,
    step=1,
    description="Year:",
    style={"description_width": "70px"},
    layout=widgets.Layout(width="400px")
)

slider_box = widgets.HBox(
    [slider_year],
    layout=widgets.Layout(justify_content="center", padding="20px 0")
)

# ===========================================================
# Precompute monthly aggregates for ALL YEARS
# ===========================================================
df_parsed = (
    df.with_columns([
        pl.col("files_oldest").alias("date"),
        pl.col("files_oldest").dt.year().alias("year"),
        pl.col("files_oldest").dt.month().alias("month")
    ])
)
years_available = sorted(df_parsed["year"].unique().to_list())
monthly_by_year = (
    df_parsed
    .group_by(["year", "month"])
    .agg([
        pl.col("files_size").sum().alias("total_size"),
        pl.col("files_count").sum().alias("total_files")
    ])
)

# ===========================================================
# Build the figure with all year traces
# ===========================================================
months_labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Create figure with dual Y-axes support
fig = make_subplots(specs=[[{"secondary_y": True}]])

for yr in years_available:
    df_year = monthly_by_year.filter(pl.col("year") == yr).sort("month")
    size_vals = [None]*12
    count_vals = [None]*12
    for row in df_year.iter_rows(named=True):
        idx = row["month"] - 1
        size_vals[idx] = row["total_size"]
        count_vals[idx] = row["total_files"]
    # ---- BAR TRACE (right axis) ----
    fig.add_trace(
        go.Bar(
            x=months_labels,
            y=count_vals,
            name=f"{yr} File Count",
            marker_color="#90caf9",
            opacity=0.5,
            visible=(yr == slider_year.value),
            hovertemplate="<b>%{x}</b><br>Files: %{y}<extra></extra>"
        ),
        secondary_y=True    # RIGHT AXIS
    )
    # ---- LINE TRACE (left axis) ----
    fig.add_trace(
        go.Scatter(
            x=months_labels,
            y=size_vals,
            mode="lines+markers",
            line=dict(color="#00bcd4", width=3),
            name=f"{yr} Data Size (TB)",
            visible=(yr == slider_year.value),
            hovertemplate="<b>%{x}</b><br>Size: %{y} TB<extra></extra>"
        ),
        secondary_y=False   # LEFT AXIS
    )

# Layout
fig.update_layout(
    height=400,
    template="plotly_white",
    barmode="overlay",
    xaxis_title="Months",
    legend_title="Legend"
)
# ---- Left Y-axis styling ----
fig.update_yaxes(
    title="Data Size (TB)",
    showgrid=True,
    gridcolor="#D0D0D0",
    gridwidth=1,
    zeroline=False,
    secondary_y=False,
    layer="above traces"
)
# ---- Right Y-axis styling ----
fig.update_yaxes(
    title="File Count",
    showgrid=True,
    gridcolor="#FFD27F",
    gridwidth=1,
    griddash="dash",
    zeroline=False,
    secondary_y=True,
    layer="below traces"
)

plot_output = widgets.Output()
with plot_output:
    fig.show()

# ===========================================================
# Slider callback (hide/show traces only)
# ===========================================================
def on_slider_change(change):
    selected_year = change["new"]
    # Each year has 2 traces: bar + line
    for i, yr in enumerate(years_available):
        bar_idx = i * 2
        line_idx = i * 2 + 1
        visible = (yr == selected_year)
        fig.data[bar_idx].visible = visible
        fig.data[line_idx].visible = visible
    # Update only visibility, no redraw flicker
    with plot_output:
        plot_output.clear_output(wait=True)
        fig.show()
slider_year.observe(on_slider_change, names="value")

# ===========================================================
# Build the dashboard tab
# ===========================================================
overview = widgets.VBox([
    cards_row,
    slider_box,
    plot_output
])
tab2 = widgets.HTML("<h3>Charts</h3><p>More content here.</p>")
tab3 = widgets.HTML("<h3>Settings</h3><p>Config panel.</p>")
tabs = widgets.Tab(children=[overview, tab2, tab3])
tabs.set_title(0, "Overview")
tabs.set_title(1, "Charts")
tabs.set_title(2, "Settings")
display(tabs)
```
